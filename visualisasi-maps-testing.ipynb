```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import folium

from warnings import simplefilter
from sklearn.exceptions import ConvergenceWarning
simplefilter ('ignore' , category="ConvergenceWarning")
```

```python
df_maps = pd.read_excel('testing.xlsx')
df_maps
```

MENAMPILKAN VISUALISASI MAP DENGAN FOLIUM 

```python
map = folium.Map(location=[df_maps.Lat.mean(), df_maps.Long.mean()], zoom_start=14, control_scale=True)
for index, location_info in df_maps.iterrows():
    folium.Marker([location_info["Lat"], location_info["Long"]], popup=location_info["City"]).add_to(map)
```
